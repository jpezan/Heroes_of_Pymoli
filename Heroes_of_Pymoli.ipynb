```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
#file_to_load = "Resources/purchase_data.csv"

# Study data files
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Display players
players = len(purchase_data['SN'].unique())
players_df = pd.DataFrame({"Total Number of Players": [len(purchase_data.SN.unique())]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Counting Number of Unique Items
items = len(purchase_data['Item ID'].unique())

#Calculate Average Price
avg_price= purchase_data['Price'].mean()

#Countint Number of Purchases
purchases= len(purchase_data['Purchase ID'])

#Calculating Total Revenue
total_rev= sum(purchase_data['Price'])

#Summary DF
summary_df = pd.DataFrame({
    'Unique items': items,
    'Average Price': '${:.2f}'.format(avg_price),
    'Total Purchases': purchases,
    'Total Revenue': '${:.2f}'.format(total_rev)
},index=[0])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Removing Duplicates from data
player_dup = purchase_data.drop_duplicates('SN')

#Total # Gender
tot_gender = player_dup['Gender'].value_counts()

#Total % Gender
tot_percent = (player_dup['Gender'].value_counts()/players)*100

#Creating new gender DF
gender_data_df = pd.DataFrame({
    'Count': tot_gender,
    'Percent of Total Players': tot_percent.map('{:.2f}%'.format)
})
gender_data_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve/Group Item ID, Item Name, and Item Price Columns
pop_items = purchase_data[['Item ID', 'Item Name', 'Price']]

count_pop_items = pop_items.groupby(["Item ID","Item Name"])["Price"].count()

avg_price_pop_items= pop_items.groupby(["Item ID","Item Name"])["Price"].mean()

tot_pop_items = pop_items.groupby(["Item ID","Item Name"])["Price"].sum()

#Popular Items DF
pop_item_df = pd.DataFrame({
    "Purchase Count":count_pop_items,
    "Average Item Price":avg_price_pop_items,
    "Total Purchase Value":tot_pop_items
})

pop_items_df = pop_item_df.sort_values("Purchase Count", ascending= False)
pop_items_df["Average Item Price"]= pop_items_df["Average Item Price"].map("${:.2f}".format)
pop_items_df["Total Purchase Value"]= pop_items_df["Total Purchase Value"].map("${:.2f}".format)
pop_items_df[:5]
```
